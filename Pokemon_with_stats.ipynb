```python
attack_mean=df['Attack'].mean()
```

```python
def set_attack(val):
  if val<attack_mean:
    return 'Attack Low'
  elif val==attack_mean:
    return 'Attack neutral'
  else:
    return 'Attack High'
```

```python
df['Attack_High_Low']=df.Attack.apply(set_attack)
```

```python
 HP_mean=df['HP'].mean()
```

```python
def  High_Low_HP(val):
  if val<HP_mean:
    return 'Low HP'
  elif val>HP_mean:
    return 'High HP'
  else:
    return 'Neutral HP'
df['High_Low_HP']=df['HP'].apply(High_Low_HP)
display(df.head())
```
